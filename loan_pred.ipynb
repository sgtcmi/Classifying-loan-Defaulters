```python
mean_earning_per_state = []
for state in training_data.STATE.unique():
    mean_earning_per_state.append([state,training_data[training_data["STATE"] == state].Income.min(),training_data[training_data["STATE"] == state].Income.mean(),training_data[training_data["STATE"] == state].Income.max()])
mean_earning_per_state = pd.DataFrame(mean_earning_per_state,
                  columns=['State', 'min_income','Average_Income','max_income'])                                  
```

```python
mean_earning_per_state.sort_values(by="Average_Income",ascending=False)
```
